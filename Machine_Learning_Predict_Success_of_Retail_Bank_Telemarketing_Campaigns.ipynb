## **Executive Summary**
 
### **Purpose**
This project analyzed the relationship between various attributes of a retail bank’s telemarketing campaign aimed at selling long-term deposits to current and prospective customers. The data was collected from 2008 to 2013, and includes 20 potential feature variables and one target variable (binary, ‘yes’ or ‘no’ for client decision to purchase a long-term deposit). The objective was to develop a Machine-Learning (ML) model that could accurately predict whether or not a client is going to purchase the financial product based on their interaction with the bank’s telemarketing campaign.  
 
### **Project Team Members**
The research/analytics team consisted of four aspiring Business Analytics professionals, all of which are currently completing their graduate degrees in the Master’s of Science - Business Analytics program within the Florida State University College of Business. 
 
### **Methods**
A standard ML project model was utilized by the project team consisting of the following stages:
Data preparation ML model selection, training/validation, testing Performance measure implementation/assessment Fine-tune the ML model

###  **Data preparation**
Prior to exposing a ML model to any data in a training environment, several steps were taken to adjust/revise the data set. All entries with missing values were dropped from the data set entirely. Prior to the creation of the training and test sets, the feature variables and target variable were isolated into new data frames. Due to the fact that half of the columns in the data set contained categorical data (several of which were binary attributes), it was also necessary to transform the data in these columns to numerical values. Additionally, the 10 columns containing numerical data had a wide range of scales which, if not addressed, could skew the results of any ML model training. In order to address these issues simultaneously, the categorical and numerical columns were isolated and a pipeline was created in order to perform both required functions simultaneously (transforming categorical data to numerical data, feature scaling for all numerical data).<br>

Additional steps were taken regarding data cleaning and/or training/validation set preparation for several of the ML models.<br>

In order to better train and assess the accuracy of the *K-Nearest Neighbors (KNN)* model, the training data set was further divided and a *validation set* was created using an approximate ⅔ / ⅓ split (training set, validation set). This was done in order to create an opportunity to assess the variance of the ML model on the training set, and by doing so, providing the research team feedback whether or not the ML model was overfitting the training data (and if so, by how much).<br>

   We implemented a neural network using Dense layers, thus requiring a 1-D tensor input. We ensured the shape of the input data and specified its dimensions in the input layer of the model. From there, keras does not require further specification of the input data in the following layers, so the remaining layers did not include an input shape.>
  
### **ML selection, training/validation, testing**
#### Logistic Regression
The training data was fit to a linear regression model using Scikit-Learn’s *LogisticRegression* algorithm, and was set to 500 iterations. After training the ML model, several performance measures were utilized (*Cross-Validation, Confusion Matrix/Classification Report, ROC Curve, Precision-Recall Curve*) in order to ascertain whether or not modifications to the parameters of the model were needed to improve overall accuracy.<br>

#### KNN
The training data and validation data was fit to a KNN ML model using Scikit-Learn’s *KNeighborsClassifier*, with initial settings of k=8. The training data and validation data were both evaluated during the execution of the eight iterations of the  *for loop*, and accuracy scores were generated for both data sets. Multiple performance measures were implemented following the training/validation of the ML model, to include a *Classification Report* and *ROC Curve*. Additionally, Scikit-Learn’s *GridSearchCV* was utilized in order to determine the best possible hyperparameters to use  for this particular data set / ML model relationship. The *GridSearchCV* was run using a range of 1-50.<br>

#### SGD 
The training data was fit to a Stochastic Gradient Descent (SGD) classifier using Scikit-Learn’s *SGDClassifier* class, with random_state=42 to guarantee reproducible results. After training the SGD model, *Cross-Validation* was used to evaluate the model. Other performance measures were implemented using cross-validated estimates of the model, to include a *Classification Report* and *ROC Curve*.<br>
#### SVM 
The training data was first fit to a Linear Support Vector Machine (SVM), with an initial setting of C=5, loss="hinge", random_state=42, max_iter=10000, multi_class='ovr' dual=True with StandardScaler. Even though the model was trained to completion, the results did not garner inclusion in this report, as the accuracy measures did not rise to the level of the competing models.  The training data was then fit to a SVM Classification model using Scikit-Learn’s built-in algorithm. After fitting the data,  several performance measures were utilized, to include a *Classification Report* and *ROC Curve.*<br>
#### NN
The neural network was created using the Sequential model from keras python library. The initial model was a simple 3-layer neural network that included an input layer, one dense hidden layer with 16 neurons, and an output layer with one neuron. We implemented the Rectified Linear Unit (Relu) as the activation for the hidden layer. This allowed the neurons to be activated when the input is greater than zero and not activated when the input is less than zero.  The output layer used sigmoid activation so the resulting output would be a probability between zero and one, thus classifying the person as 1 (will subscribe) or 0 (will not subscribe). Since this is a binary output, the last layer has only one neuron. Lastly, the network was compiled using an adam optimizer , binary cross entropy loss function, and ROC AUC score. The model was improved after the initial results, see results below for further explanation.<br>

#### *Analysis* 
Due to the fact that the data set was heavily skewed in favor of instances where a potential client did not purchase the financial product (i.e., a “No” or 0), it was difficult to ascertain the effectiveness of the various ML models using certain metrics. As shown in the model comparison table above, accuracy scores for the models ranged from 0.77 to 0.91. However, upon review of more direct metrics such as precision, recall, and f1 scores, it is apparent that the accuracy scores benefited from the fact that close to 90% of all training instances resulted in a “No” (0) for customer response to the marketing effort. For this reason, the discrepancy between the precision, recall, and f1 scores for each model varied wildly between that of training instances involving positive (1) and negative (0) outcomes. In order to more clearly distinguish between the various ML models, a ROC AUC score was generated for each model showing the discrepancy between the True Positive Rate (TPR) and the False Positive Rate (FPR). Based on the uneven nature of training instances involving positive (1) and negative (0) outcomes, this metric was needed to determine which ML model performed the best. Despite the fact that the Logistic Regression model is the least complex among the models used in this research, it outperformed the others and produced the highest ROC AUC score (0.79, edging out the Neural Net slightly).
 
### **Recommendations**
A data set that was more balanced between each binary class in the target variable would be more likely to produce a ML model with greater accuracy. That being said, short of manufacturing the data, it is unrealistic to expect an even split between customers who buy a particular product and those who don’t. This is simply not how sales campaigns (particularly telemarketing campaigns) success ratios work out in most cases. There will likely always be a significantly higher amount of “No” outcomes in sales campaigns like this, but if data was collected over a longer period of time, it is possible to increase the balance between in the target variable classes. Any improvement in this ratio would likely serve to improve the accuracy of a trained ML model. 


# <font color=#DD4124>Preparing the Data</font>

```python
#import libraries
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow as tf

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_curve
from sklearn.svm import SVC 
from sklearn.pipeline import make_pipeline
from keras import models
from keras import layers
from sklearn.model_selection import GridSearchCV
```

```python
#read in data
data = pd.read_csv('banking.csv')
data.head()
```

```python
#Handle missing values in this dataset if they exist.
data = data.dropna()
```

```python
# Let's see if it worked:
data.isnull().sum()

# In the output below, we should see '0' missing values for each of the variables.
```

```python
#seperate x and y values 
X = data.drop(['y', 'duration'],axis=1) #see note below
y = data['y']
print(X.shape)
print(y.shape)
```

Duration Note:
1. The feature variable was removed from the entire dataset because the information provided in the project description. Since duration is not determined until after the phone call and would not be helpful to predict subscription to the bank, we removed it from the data set.

```python
#split train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size =.8, test_size=.2,
                                                    random_state=42, stratify=y)

#split data to have 80/20 train/ test split
#we stratified data by output feature variable 'y' to ensure even spread of data in both data sets
```

```python
#let's visualize the data
X_train.hist(figsize=(10,15));
```

```python
#now that we have an idea of what the numerical data looks like, let's process the observations
#first, we will seperate the categorical and numerical data
cat_attribute = X_train.drop(['age', 'campaign', 'pdays', 'previous',
                              'emp_var_rate', 'cons_price_idx', 'cons_conf_idx',
                             'euribor3m', 'nr_employed'], axis=1)
#using the description, we isolated the categorical values and numerical values
#we will modify these variables by placing them through a column transformer
```

```python
#quick for-loop to isolate the numerical variables
#isolate the the num col names
num_col = []
for col in list(X_train):
    if col not in list(cat_attribute):
        num_col.append(col)
#num col will have all numerical variables except for duration, which was removed at the beginning of data processing
```

```python
#We will use the column transformer to One-Hot Categorical variables and standardize numerical variables
#use column transformer
cat_col = list(cat_attribute)

full_pipeline = ColumnTransformer([
    ('cat', OneHotEncoder(sparse=False), cat_col),
    ('num', StandardScaler(), num_col),
])

X_train_prepared = full_pipeline.fit_transform(X_train)
X_train_prepared.shape #check shape
#all categorical variables were transformed to one-hots, resulting in a dataset of 32950 by 62
```

# <font color=#DD4124>Selecting and Training the Model</font>

In this section, we will train and evaluate a Logistic Regrssion model, SGD Classifier, Support Vector Machine, and a Neural Network using the input data we prepared.

## <font color=#5B5EA6>Logistic Regression</font>

```python
#fit data to logistic regression model from scikit learn
lr = LogisticRegression(max_iter=500,random_state=0)
lr.fit(X_train_prepared, y_train)

lr_score = lr.score(X_train_prepared, y_train)
print('Accuracy of {:.2%} with basic lr model'.format(lr_score))
```

```python
# Series of performance measures to see if we can improve the accuracy of the model
lr_scores = cross_val_score(lr, X_train_prepared, y_train, cv=3, scoring="accuracy")
#create a function to display scores
def display_scores(scores):
    print("Scores:",scores)
    print("Mean: {:.2f}".format(scores.mean()))
    print ("Standard deviation: {:.2f}".format(scores.std()))
    
display_scores(lr_scores)

#the scores are relatively the same as results from the training set, the data does not seem to be overiftting
```

```python
#lets use the classification report to find the true performance of this classifier
lr_pred= cross_val_predict(lr, X_train_prepared, y_train, cv=3)

print(classification_report(y_train, lr_pred))
#we can see the model is very good at predicting when the customer won't subscribe but poor at determining when
#they will subscribe
# Worse than a coin flip when the ML model encouters '1's 
```

```python
#As a final measure of performance, we will use the aoc_auc_score
#use sklearns roc_curve to return fpr, tpr, and thresholds to plot
fpr, tpr, thresholds = roc_curve(y_train, lr_pred)

plt.style.use('ggplot')
def plot_roc_curve(fpr, tpr, label=None):
    plt.title("ROC Curve", fontsize=25)
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.grid(True) #shows grid lines
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)

plt.figure(figsize=(8, 6))
plot_roc_curve(fpr, tpr)
plt.show()
```

### <font color=#9B2335>Results</font>

```python
lr_pred= cross_val_predict(lr, X_train_prepared, y_train, cv=3,method="decision_function")
roc_auc_score(y_train, lr_pred)
```

The logistic regression showed a lot about the data. The model was great at predicting customers that 'will not'(0) subscribe but did not perform as well predicting customers that 'will'(1) subscribe. The auc score will be used to compare the performance to other models.

## <font color=#5B5EA6>K-Nearest Neighbors</font>

```python
X_val = X_train_prepared[:10000]
partial_X_train_prepared = X_train_prepared[10000:]

y_val = y_train[:10000]
partial_y_train = y_train[10000:]

# Setting aside a validation set (from the training set) to use for this model
```

```python
neighbors = np.arange(1,9)
train_accuracy =np.empty(len(neighbors))
validation_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)

    #Fit the model
    knn.fit(partial_X_train_prepared, partial_y_train)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(partial_X_train_prepared, partial_y_train)
    
    #Compute accuracy on the test set
    validation_accuracy[i] = knn.score(X_val, y_val) 

print('Training accuracy is: ',knn.score(partial_X_train_prepared, partial_y_train))
print('Validation accuracy is: ',knn.score(X_val, y_val)) 

# Runing model with k=8 for initial pass...no better than lr at this point...slight overfitting of training vs validation set as well 
```

```python
#let's visualize the accuracy results
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, validation_accuracy, label='Validation Accuracy')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```

We can see the model has high variance at first but decreases as the number of neighbors increase

```python
#let's see the classification report
knn_pred = knn.predict(X_val)
print(classification_report(y_val, knn_pred))
```

```python
y_pred_proba = knn.predict_proba(X_val)[:,1]

fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)

plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr, label='Knn')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('Knn(n_neighbors=7) ROC curve')
plt.show()
```

```python
#let's see the best parameters of the k nearest neighbors since we only went to 8 k
param_grid = {'n_neighbors':np.arange(1,50)}

knn = KNeighborsClassifier()
knn_cv= GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(X_train_prepared, y_train)

# Using GridSearch to find optimal parameters  (just a heads up, running this took a long time)
```

```python
def precision_recall_curve(precisions, recalls, thresholds):
    fig, ax = plt.subplots(figsize=(12,8))
    plt.plot(thresholds, precisions[:-1], "r--", label="Precisions")
    plt.plot(thresholds, recalls[:-1], "#424242", label="Recalls")
    plt.title("Precision and Recall \n Tradeoff", fontsize=18)
    plt.ylabel("Level of Precision and Recall", fontsize=16)
    plt.xlabel("Thresholds", fontsize=16)
    plt.legend(loc="best", fontsize=14)
    plt.xlim([-2, 4.7])
    plt.ylim([0, 1])
    plt.axvline(x=-1.04, linewidth=2, color="#0B3861")
    plt.annotate('Best Precision and \n Recall Balance \n is at -1.04 \n threshold ', xy=(-1.04, 0.43), xytext=(55, -40),
             textcoords="offset points",
            arrowprops=dict(facecolor='black', shrink=0.05),
                fontsize=12, 
                color='k')
    
precision_recall_curve(precisions, recalls, threshold)
plt.show()
```

## <font color=#5B5EA6>Neural Network</font>

```python
#Let's see how well the neural network performs this classification task
#we will start with a basic 3 layer neural network with 16 neurons
nw = models.Sequential()

#add layers to model
nw.add(layers.Dense(16, activation='relu', input_shape=(62,)))
nw.add(layers.Dense(1, activation='sigmoid'))
nw.summary()
```

```python
#since keras doesn't have roc built into it as a performance measure,
#we will use this quick piece of code found at #https://stackoverflow.com/questions/41032551/how-to-compute-receiving-operating-characteristic-roc-and-auc-in-keras
#the following function returns the roc auc score
def auroc(y_true, y_pred):
    return tf.py_function(roc_auc_score, (y_true, y_pred), tf.double)

#compile model using adam for the optimizer and binary_crossentropy as the loss function
nw.compile(optimizer='adam',
          loss='binary_crossentropy',#using binary cross entropy since y is binary
          metrics=['accuracy', auroc]) 
```
